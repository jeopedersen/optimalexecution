```python
import numpy as np
import matplotlib.pyplot as plt
import random as random

from seaborn import color_palette, set_palette
from matplotlib import rc, cm

plt.style.use('seaborn-white')
set_palette("deep")

rc('font', **{'family': 'serif', 'serif': ['Palatino']})
rc('text', usetex=True)
rc('text.latex', preamble=r'\usepackage{underscore}')

prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']
```

```python
### Perold's Implementation Shortfall Example
years = np.arange(1966, 1987)
# Paper portfolio

factor = 1.20  # Define the factor for multiplication
iterations = 20  # Define the number of iterations

array_paper = [100]  # Start with the first value as 100

for _ in range(iterations):
    previous_value = array_paper[-1]  # Get the previous value from the array
    new_value = previous_value * factor  # Calculate the new value
    array_paper.append(new_value)  # Append the new value to the array

print(array_paper)

```

```python
### Perold's Implementation Shortfall Example

# Actual portfolio

factor = 1.025  # Define the factor for multiplication
iterations = 20  # Define the number of iterations

array_actual = [100]  # Start with the first value as 100

for _ in range(iterations):
    previous_value = array_actual[-1]  # Get the previous value from the array
    new_value = previous_value * factor  # Calculate the new value
    array_actual.append(new_value)  # Append the new value to the array

print(array_actual)
```

```python
plt.figure(figsize=(6,5), dpi=100)
# Figure 1
plt.plot(years, array_actual, marker='o', lw=2.0, label='Actual')
plt.plot(years, array_paper, marker='o', lw=2.0, label='Paper')
plt.grid(True)
plt.xlabel(r'Year')
plt.ylabel(r'Portfolio Value')
plt.xticks(years[::5])
plt.title('Value Line Fund (Actual vs. Paper)')
plt.savefig('perold.pdf')
plt.legend(loc='best',frameon=True)
```
